```python
# Fast Services Corporation - KPI, KRI & Sales Dashboard
# ================================================

# Cell 1: Install required libraries
!pip install plotly pandas numpy

import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

print("✅ Libraries installed successfully!")

# Cell 2: Mount Google Drive (Your code)
from google.colab import drive
drive.mount('/content/drive')

# Cell 3: Create project folder
%cd /content/drive/MyDrive/
!mkdir -p Fast_ServicesDashboard
%cd Fast_ServicesDashboard
!pwd

# Cell 4: Generate Realistic Dataset
print("\n🔄 Generating realistic dataset...")

# Set random seed for reproducibility
np.random.seed(42)

# Product categories and items (Fast Services Corporation products)
product_data = {
    'Sanitary Products': ['Sanitary Pads Regular', 'Sanitary Pads Overnight', 'Pantyliners'],
    'Baby Care': ['Baby Diapers S', 'Baby Diapers M', 'Baby Diapers L', 'Baby Wipes'],
    'Paper Products': ['Facial Tissues', 'Paper Towels', 'Toilet Paper'],
    'Personal Care': ['Wet Wipes Adult', 'Hand Sanitizer'],
    'Household Items': ['Paper Napkins', 'Kitchen Rolls']
}

# Generate realistic monthly revenue with seasonal patterns
months = pd.date_range('2024-01-01', '2024-12-01', freq='MS')
months_str = months.strftime('%b %Y')

# Base revenue with seasonal adjustments
# Higher in March (women's month), November-December (holiday season)
# Lower in January, February, August (lean months)
base_revenue = [1200000, 950000, 1500000, 1300000, 1100000, 1600000,
                1550000, 1050000, 1350000, 1600000, 2200000, 1800000]
monthly_revenue = [r + np.random.uniform(-150000, 150000) for r in base_revenue]

# Create transaction dataset
transactions = []
customer_ids = [f'Customer_{str(i).zfill(3)}' for i in range(1, 51)]

order_id = 1001
for month_idx, month in enumerate(months):
    # Number of orders varies by month (correlated with revenue)
    num_orders = int(monthly_revenue[month_idx] / 15000) + np.random.randint(-5, 5)

    for _ in range(num_orders):
        # Random product selection weighted by category popularity
        category = np.random.choice(list(product_data.keys()),
                                   p=[0.25, 0.30, 0.20, 0.15, 0.10])
        product = np.random.choice(product_data[category])

        # Price varies by product type
        if 'Diaper' in product:
            unit_price = np.random.uniform(180, 350)
        elif 'Sanitary' in product or 'Pad' in product:
            unit_price = np.random.uniform(80, 200)
        elif 'Tissue' in product or 'Paper' in product:
            unit_price = np.random.uniform(30, 100)
        else:
            unit_price = np.random.uniform(50, 180)

        quantity = np.random.randint(1, 20)
        total_amount = unit_price * quantity

        # Random date within the month
        days_in_month = (month + pd.offsets.MonthEnd(0) - month).days
        order_date = month + timedelta(days=np.random.randint(0, days_in_month))

        transactions.append({
            'Order Id': f'ORD-{order_id}',
            'Order Date': order_date.strftime('%Y-%m-%d'),
            'Customer Name': np.random.choice(customer_ids),
            'Category': category,
            'Product Name': product,
            'Quantity': int(quantity),
            'Unit Price': round(unit_price, 2),
            'Total Amount': round(total_amount, 2)
        })
        order_id += 1

transactions_df = pd.DataFrame(transactions)
print(f"✅ Generated {len(transactions_df)} transactions")

# Cell 5: Create Dashboard Visualizations
print("\n📊 Creating dashboard visualizations...")

# Summary metrics
total_revenue = transactions_df['Total Amount'].sum()
total_orders = len(transactions_df)
avg_order_value = total_revenue / total_orders
total_units = transactions_df['Quantity'].sum()

# Monthly revenue aggregation
monthly_rev_df = transactions_df.groupby(transactions_df['Order Date'].str[:7]).agg({
    'Total Amount': 'sum',
    'Order Id': 'count'
}).reset_index()
monthly_rev_df.columns = ['Month', 'Revenue', 'Orders']
monthly_rev_df = monthly_rev_df.sort_values('Month')

# Category revenue
category_rev = transactions_df.groupby('Category')['Total Amount'].sum().sort_values(ascending=False)

# Create dashboard layout using Plotly
fig_dashboard = make_subplots(
    rows=3, cols=2,
    subplot_titles=('Monthly Revenue Trend', '',
                    'Revenue by Category', 'Orders by Month',
                    '', ''),
    specs=[[{"colspan": 2, "secondary_y": False}, None],
           [{"type": "bar"}, {"type": "bar"}],
           [{"colspan": 2, "type": "table"}, None]],
    vertical_spacing=0.10,
    horizontal_spacing=0.08,
    row_heights=[0.35, 0.30, 0.35]
)

# 1. Monthly Revenue Trend (Full width)
fig_dashboard.add_trace(
    go.Scatter(
        x=monthly_rev_df['Month'],
        y=monthly_rev_df['Revenue'],
        mode='lines+markers',
        fill='tozeroy',
        fillcolor='rgba(0, 100, 180, 0.15)',
        line=dict(color='#0064B4', width=3),
        marker=dict(size=8, color='#0064B4', line=dict(width=2, color='white')),
        name='Revenue',
        hovertemplate='<b>%{x}</b><br>Revenue: ₱%{y:,.0f}<extra></extra>'
    ),
    row=1, col=1
)

# 2. Revenue by Category (Horizontal Bar)
fig_dashboard.add_trace(
    go.Bar(
        x=category_rev.values,
        y=category_rev.index,
        orientation='h',
        marker=dict(color=['#002855', '#00509E', '#0066CC', '#4DA6FF', '#80BFFF']),
        text=[f'₱{v:,.0f}' for v in category_rev.values],
        textposition='outside',
        name='Category',
        hovertemplate='<b>%{y}</b><br>Revenue: ₱%{x:,.0f}<extra></extra>'
    ),
    row=2, col=1
)

# 3. Orders by Month
fig_dashboard.add_trace(
    go.Bar(
        x=monthly_rev_df['Month'],
        y=monthly_rev_df['Orders'],
        marker=dict(color='#0064B4', opacity=0.8),
        name='Orders',
        hovertemplate='<b>%{x}</b><br>Orders: %{y}<extra></extra>'
    ),
    row=2, col=2
)

# 4. Transaction Details Table (First 25 rows)
# Prepare table data
table_data = transactions_df.head(25)
cell_values = [
    table_data['Order Id'].tolist(),
    table_data['Order Date'].tolist(),
    table_data['Customer Name'].tolist(),
    table_data['Category'].tolist(),
    table_data['Product Name'].tolist(),
    table_data['Quantity'].tolist(),
    [f"₱{price:,.2f}" for price in table_data['Unit Price'].tolist()],
    [f"₱{amount:,.2f}" for amount in table_data['Total Amount'].tolist()]
]

fig_dashboard.add_trace(
    go.Table(
        header=dict(
            values=['<b>Order ID</b>', '<b>Date</b>', '<b>Customer</b>',
                    '<b>Category</b>', '<b>Product</b>', '<b>Qty</b>',
                    '<b>Unit Price</b>', '<b>Total</b>'],
            fill_color='#002855',
            align='left',
            font=dict(color='white', size=12),
            height=30
        ),
        cells=dict(
            values=cell_values,
            fill_color=[['#fafafa', '#f0f4f8']*12],
            align='left',
            font=dict(size=10),
            height=25
        )
    ),
    row=3, col=1
)

# Update layout
fig_dashboard.update_layout(
    height=1300,
    width=1100,
    title_text='<b>FAST SERVICES CORPORATION - Sales Dashboard</b>',
    title_font_size=26,
    title_font_color='#002855',
    showlegend=False,
    template='plotly_white',
    margin=dict(l=80, r=50, t=100, b=50),
    hovermode='x unified'
)

# Format axes
fig_dashboard.update_yaxes(title_text='<b>Revenue (PHP)</b>', row=1, col=1,
                          tickformat=',.0f', gridcolor='lightgray')
fig_dashboard.update_xaxes(tickangle=-45, row=1, col=1, gridcolor='lightgray')

fig_dashboard.update_yaxes(title_text='<b>Revenue (PHP)</b>', row=2, col=1,
                          tickformat=',.0f', gridcolor='lightgray')
fig_dashboard.update_xaxes(title_text='<b>Category</b>', row=2, col=1)

fig_dashboard.update_yaxes(title_text='<b>Number of Orders</b>', row=2, col=2,
                          gridcolor='lightgray')
fig_dashboard.update_xaxes(tickangle=-45, row=2, col=2, gridcolor='lightgray')

# Show dashboard
fig_dashboard.show()

# Cell 6: Save dashboard to Google Drive
print("\n💾 Saving dashboard to Google Drive...")
fig_dashboard.write_html('/content/drive/MyDrive/Fast_ServicesDashboard/Sales_Dashboard.html')
print("✅ Dashboard saved successfully!")

# Cell 7: Performance Summary
print("\n" + "="*80)
print("📈 FAST SERVICES CORPORATION - PERFORMANCE SUMMARY (2024)")
print("="*80)
print(f"\n💰 Total Revenue:           ₱{total_revenue:,.2f}")
print(f"📦 Total Orders:             {total_orders:,}")
print(f"📊 Average Order Value:      ₱{avg_order_value:,.2f}")
print(f"📦 Total Units Sold:         {total_units:,}")
print("\n🏆 Top Product Category:")
print(f"   {category_rev.index[0]} - ₱{category_rev.values[0]:,.2f}")
print("\n📅 Best Performing Month:")
best_month = monthly_rev_df.loc[monthly_rev_df['Revenue'].idxmax()]
print(f"   {best_month['Month']} - ₱{best_month['Revenue']:,.2f} ({best_month['Orders']} orders)")
print("="*80)

# Cell 8: Show sample transactions
print("\n📋 SAMPLE TRANSACTION DETAILS (First 15 records):")
print("-"*80)
print(transactions_df.head(15).to_string(index=False))
```
