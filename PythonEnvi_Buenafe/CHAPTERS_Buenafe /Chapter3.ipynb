```python
hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

if hours <= 40:
    pay = hours * rate
else:
    pay = 40 * rate + (hours - 40) * rate * 1.5

print(f"Pay: {pay}")
```

```python
def computegrade(score):
    """Takes a score and returns the corresponding grade string."""
    if 97 <= score <= 100:
        return "1.0"
    elif 93 <= score < 97:
        return "1.25"
    elif 89 <= score < 93:
        return "1.5"
    elif 85 <= score < 89:
        return "1.75"
    elif 81 <= score < 85:
        return "2.0"
    elif 77 <= score < 81:
        return "2.25"
    elif 74 <= score < 77:
        return "2.5"
    elif 0 <= score < 74:
        return "3.0"
    else:
        return "Invalid score"

# Run the program repeatedly
print("=== Grade Calculator ===")
while True:
    user_input = input("\nEnter a score (or 'q' to quit): ")

    # Check if user wants to quit
    if user_input.lower() == 'q':
        print("Goodbye!")
        break

    try:
        score = float(user_input)
        grade = computegrade(score)
        print(f"Score: {score} → Grade: {grade}")
    except ValueError:
        print("Invalid input! Please enter a number.")
```
