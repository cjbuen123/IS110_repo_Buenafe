```python
def compute_pay(hours, rate): #correct putput
    """Calculate pay with time-and-a-half for overtime."""
    if hours <= 40:
        # No overtime
        pay = hours * rate
    else:
        # Regular pay for first 40 hours
        regular_pay = 40 * rate
        # Overtime pay for hours over 40 (1.5x rate)
        overtime_pay = (hours - 40) * rate * 1.5
        pay = regular_pay + overtime_pay

    return pay

# Get user input
hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

# Calculate and display pay
pay = compute_pay(hours, rate)
print(f"Pay: {pay}")
```

```python
# My Output
while True:
    try:
        score = float(input("Enter score (or -1 to quit): "))

        if score == -1:
            break

        # Validate score range
        if score < 0 or score > 100:
            print("Invalid score")
        elif 97 <= score <= 100:
            print("Grade: 1.0")
        elif 93 <= score < 97:
            print("Grade: 1.25")
        elif 89 <= score < 93:
            print("Grade: 1.5")
        elif 85 <= score < 89:
            print("Grade: 1.75")
        elif 81 <= score < 85:
            print("Grade: 2.0")
        elif 77 <= score < 81:
            print("Grade: 2.25")
        elif 74 <= score < 77:
            print("Grade: 2.5")
        elif 0 <= score < 74:
            print("Grade: 3.0")
        else:
            print("Invalid score")

    except ValueError:
        print("Invalid input. Please enter a numeric score.")
```
